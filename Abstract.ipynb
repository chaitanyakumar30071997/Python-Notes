#                                                          Abstract

1. Abstract class:-  If a class contains at least one Abstract method.
2. Abstract Method :- A method without definition.
3. Concrete classes: - If a class doesn't contain any Abstract method.
4.  Abstraction:- Abstraction is the mechamisn of only declaring method but not instantiated Declased method are implemented in its subclass .
5.  We cannot create an object for  Abstract class . But we can create an object for concreate class .
6.  To convert  Abstract class into concreate class using inheritance concept.
7.  To implementing  Abstract class and  Abstract method we need to impot module ABC
8.  ABC works by decorating methods of the base class as an abstract and then registering concrete classes as implementations of the abstract base. A method becomes abstract when decorated with the keyword @abstractmethod.

```python
from abc import ABC, abstractmethod
class Robo(ABC):   #Abstract base class  
    @abstractmethod
    def sana(self):
        print("Hello")

class chitti(Robo): # concreate class 
    def sana (self):
        print("Hai")

obj = chitti() 
obj.sana()
```

```python
from abc import ABC, abstractmethod

# Base abstract class
class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

    def move(self):
        print("I can move")

# Intermediate abstract class
class Mammal(Animal):
    @abstractmethod
    def habitat(self):
        pass

# Concrete class
class Dog(Mammal):
    def sound(self):
        print("Bark")

    def habitat(self):
        print("Domestic")

# Another concrete class
class Cat(Mammal):
    def sound(self):
        print("Meow")

    def habitat(self):
        print("Domestic")

# Creating instances of Dog and Cat
dog = Dog()
cat = Cat()

dog.sound()    
dog.habitat() 
dog.move()    

cat.sound()   
cat.habitat() 
cat.move()    
```

```python
from abc import ABC, abstractmethod

# Base abstract class
class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

    @abstractmethod
    def stop_engine(self):
        pass

    @abstractmethod
    def Auto_Mode(self):
        pass

# Intermediate abstract class
class Car(Vehicle):
    @abstractmethod
    def open_doors(self):
        pass

# Concrete class
class Sedan(Car):
    def start_engine(self):
        print("Sedan engine started")
    
    def Auto_Mode(self):
        print("Sedan AUTO MODE ON")

    def stop_engine(self):
        print("Sedan engine stopped")

    def open_doors(self):
        print("Sedan doors opened")

# Another concrete class
class SUV(Car):
    def start_engine(self):
        print("SUV engine started")
    
    def Auto_Mode(self):
        print("SUV Auto Mode ON")

    def stop_engine(self):
        print("SUV engine stopped")

    def open_doors(self):
        print("SUV doors opened")

class BMW(Car):
    def start_engine(self):
        print("BMW engine started")
    
    def Auto_Mode(self):
        print("BMW Auto Mode ON")

    def stop_engine(self):
        print("BMW engine stopped")

    def open_doors(self):
        print("BMW doors opened")

# Creating instances of Sedan and SUV
sedan = Sedan()
suv = SUV()
bmw = BMW()

sedan.start_engine()  
sedan.open_doors()   
sedan.stop_engine() 
sedan.Auto_Mode() 

suv.start_engine()   
suv.open_doors()     
suv.stop_engine()  
suv.Auto_Mode()  

bmw.start_engine()  
bmw.open_doors()   
bmw.stop_engine() 
bmw.Auto_Mode() 
```
